```python
# Storage paths for the Crops and the Panos
pano_file_directory = "/Users/marcus/Desktop/Training_Data/Pano_4096_2048/" 
crop_file_directory = "/Users/marcus/Desktop/Training_Data/Crops_From_4096_2048/"
csv_out_file_directory = "/Users/marcus/Desktop/Training_Data/CSV_Output"
```

```python
import os
import numpy as np
from skimage import io
from skimage.feature import match_template
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
import json
import math
import csv
```

```python
# Parses through the pano json file, and creates Pano Objects 

pano_object_list = []

for dir_name, sub_dir_list, file_list in os.walk(pano_file_directory):
    for file_name in file_list:
        if file_name[-5:] == ".json" and len(file_name) > 20:
            
            new_pano_object = pano_object(file_name, "_4096_2048")
            pano_object_list.append(new_pano_object)
```

```python
class pano_object():
    
    pano_image_width = 4096
    pano_image_height = 2048
    
    
    def __init__(self, json_file, size_extension = ""):
        self.name = json_file[:-21]
        self.json_path = os.path.join(pano_file_directory, self.name + size_extension + "_label.json")
        self.pano_path = os.path.join(pano_file_directory, self.name + size_extension + ".jpg")
        self.labels = []
        
        with open(self.json_path, 'r') as pano_json:
            label_data = json.load(pano_json)
            for label in label_data[self.name]:
                new_label = label_object(label, self.name)
                self.labels.append(new_label)
        self.label_size = len(self.labels)
    
    def get_pano_image(self):
        pano = io.imread(self.pano_path)[:,:,2]
        return pano
    
    
class label_object():
    
    def __init__(self, label_data, pano_name):
        self.id = label_data['label_id']
        self.type = label_data['label_type']
        self.x = label_data['x']
        self.y = label_data['y']
        self.pano_name = pano_name
        crop_file_name = "{}_._{}_._{}_._{}_._.jpg".format(pano_name, label_data['label_type'], label_data['x'], label_data['y'] )
        self.label_crop_path = os.path.join(crop_file_directory, label_data['label_type'], pano_name[:2], crop_file_name)
        
    def get_label_crop_image(self):
        crop = io.imread(self.label_crop_path)[:,:,2]
        return crop
    
        
```

```python
def compare_labels(test_label, train_label):
    if test_label.type == train_label.type:
        return True
    else:
        return False

def get_distance(label, x, y):
    x1 = min(label.x, x)
    x2 = max(label.x, x)

    dx = min((x2 - x1), (pano_object.pano_image_width - x2 + x1))
    dy = label.y - y

    distance = math.sqrt((dx)**2 + (dy)**2)

    return distance

def get_total_operations(test, train):
    test_label_amt = 0
    for test_pano in test:
        test_label_amt += test_pano.label_size
    train_label_amt = 0
    for train_pano in train:
        train_label_amt += train_pano.label_size   
        
    total = test_label_amt * train_label_amt
    return total
    
```

```python
data = np.array(pano_object_list, dtype = np.object)

number_of_folds = 2
randomize = True


kfold = KFold(number_of_folds, randomize, 4)

k = 0
for train, test in kfold.split(data):
    total_operations = get_total_operations(data[train], data[test])
    csv_file_name = "distance_type_calculation_" + str(k) + ".txt"
    csv_file_path = os.path.join(csv_out_file_directory, csv_file_name)
    k += 1
    status_count = 0
    with open(csv_file_path, 'w') as csvfile:
        field_names = ['target_pano_id', 'test_pano_id', 'target_label_id', 'test_label_id', 'target_x', 'target_y', 'test_x_pred', 'test_y_pred', 'target_type', 'test_type', 'distance', 'is_same_label_type', 'threshold' ]
        writer = csv.DictWriter(csvfile, fieldnames = field_names)
        writer.writeheader()
        for train_pano in data[train]:
            
            train_pano_image = train_pano.get_pano_image()
            for test_pano in data[test]:
                for test_label in test_pano.labels:
                    
                    template = test_label.get_label_crop_image()
                    result = match_template(train_pano_image, template)
                    threshold = np.argmax(result)
                    x_pred, y_pred = np.unravel_index(threshold, result.shape)[::-1]

                    for train_label in train_pano.labels:
                        is_same_label = compare_labels(train_label, test_label)
                        distance = get_distance(train_label, x_pred, y_pred)
                        row_data = {
                            'target_pano_id' : train_pano.name,
                            'test_pano_id'   : test_pano.name,
                            'target_label_id': train_label.id,
                            'test_label_id'  : test_label.id,
                            'target_x'       : train_label.x,
                            'target_y'       : train_label.y,
                            'test_x_pred'    : x_pred,
                            'test_y_pred'    : y_pred,
                            'target_type'    : train_label.type,
                            'test_type'      : test_label.type,
                            'distance'       : distance,
                            'is_same_label_type'   : is_same_label,
                            'threshold'      : threshold
                        }
                        writer.writerow(row_data)
                        status_count += 1
                        print("Status: {} of {} label comparisons for {} of {} folds.".format(status_count, total_operations, k, number_of_folds), end = '\r')
                        
                    

                        
```

```python
def get_overlap(x1, y1, x2, y2, crop_size = 246):
    dx = max(0, (crop_size - abs(x1 - x2)))
    dy = max(0, (crop_size - abs(y1 - y2)))
    crop_area = crop_size**2
    overlap_area = max(0, (dx * dy))
    
    percentage = (overlap_area / crop_area)
    
    return percentage
```

```python
# Reads through the CSV file for Data Analysis 

csv_file_directory = "/Users/marcus/Desktop/Training_Data/CSV_Output/first_test_files_8_30_2018/"

csv_file_paths = []
distance_data = []
true_distance_data = []
false_distance_data = []
overlap_data = []
true_overlap = []
false_overlap = []

type_match_min_distance = []

for dir_name, sub_dir_list, file_list in os.walk(csv_file_directory):
    for file_name in file_list:
        if file_name[-4:] == ".csv" and len(file_name) > 20: 
            csv_file_path = os.path.join(dir_name, file_name)
            csv_file_paths.append(csv_file_path)

old_pano_id = ""
for file_path in csv_file_paths:
    
    csv_f = open(file_path)
    csv_file = csv.reader(csv_f)
    skip_first_row = True
    for row in csv_file:
        if(skip_first_row):
            skip_first_row = False
            continue
        pano_id = row[0]
        distance = float(row[10])
        target_x = float(row[4])
        target_y = float(row[5])
        test_x   = float(row[6])
        test_y   = float(row[7])
        
        distance_data.append(distance)
        overlap = get_overlap(target_x, target_y, test_x, test_y)
        overlap_data.append(overlap)
        if row[11][0] == 'T':
            true_distance_data.append(distance)
            true_overlap.append(overlap)
            if old_pano_id != pano_id:
                type_match_min_distance.append(distance)
                old_pano_id = pano_id
            else:
                index = len(type_match_min_distance) - 1
                old_dist = type_match_min_distance[index]
                type_match_min_distance[index] = min(old_dist, distance)
        else:
            false_distance_data.append(distance)
            false_overlap.append(overlap)
            
        
        
distance_data = np.array(distance_data)
true_distance_data = np.array(true_distance_data)
false_distance_data = np.array(false_distance_data)

type_match_min_distance = np.array(type_match_min_distance)

overlap_data  = np.array(overlap_data)
true_overlap  = np.array(true_overlap)
false_overlap = np.array(false_overlap)
```

```python
fig = plt.figure(figsize = (15, 15))
ax1 = plt.subplot(5,3,1)
ax2 = plt.subplot(5,3,2)
ax3 = plt.subplot(5,3,3)
ax4 = plt.subplot(5,3,7)
ax5 = plt.subplot(5,3,8)
ax6 = plt.subplot(5,3,9)

ax7 = plt.subplot(5,3,14)

ax1.hist(distance_data, bins = 500)
ax1.set_title("All Distances")
ax1.set_xlabel("Distance in pixels \n Median: {} \n Average: {} \n SD: {}".format(np.median(distance_data), distance_data.mean(), distance_data.std()))

ax2.hist(true_distance_data, bins = 500)
ax2.set_title("Distance w/ Matching Labels")
ax2.set_xlabel("Distance in pixels \n Median: {} \n Average: {} \n SD: {}".format(np.median(true_distance_data), true_distance_data.mean(), true_distance_data.std()))

ax3.hist(false_distance_data, bins = 500)
ax3.set_title("Distance w/o Matching Labels")
ax3.set_xlabel("Distance in pixels \n Median: {} \n Average: {} \n SD: {}".format(np.median(false_distance_data), false_distance_data.mean(), false_distance_data.std()))

ax4.hist(overlap_data, bins = 10)
ax4.set_title("Overlap w/ All Data")
ax4.set_xlabel("Overlap Percent \n Median: {} \n Average: {} \n SD: {}".format(np.median(overlap_data), overlap_data.mean(), overlap_data.std()))

ax5.hist(true_overlap, bins = 10)
ax5.set_title("Overlap w/ Matching Labels")
ax5.set_xlabel("Overlap Percent \n Median: {} \n Average: {} \n SD: {}".format(np.median(true_overlap), true_overlap.mean(), true_overlap.std()))

ax6.hist(false_overlap, bins = 10)
ax6.set_title("Overlap w/o Matching Labels")
ax6.set_xlabel("Overlap Percent \n Median: {} \n Average: {} \n SD: {}".format(np.median(false_overlap), false_overlap.mean(), false_overlap.std()))

ax7.hist(type_match_min_distance, bins = 20)
ax7.set_title("Closest Same-Type Template Match Distance")
ax7.set_xlabel("Distance \n Median: {} \n Average: {} \n SD: {}".format(np.median(type_match_min_distance), type_match_min_distance.mean(), type_match_min_distance.std()))

plt.show()
```
